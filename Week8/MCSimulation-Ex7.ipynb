#### (a)Write a program that uses the Rejection Sampling Method to generate a set of 10,000
#### samples from this distribution. 

```python
N = 10000
#http://en.wikipedia.org/wiki/Rejection_sampling
f = lambda x: 60 * x ** 3 * (1 - x) ** 2
#F = lambda x: 
x = arange (0.,1.01,.01)
xmin = 0.
xmax = 0.6
ymax= 2.07
ymin = 0.
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);
x_samples = []
y_samples = []
for i in range(len(samples_x)):
    if samples_y[i]<=f(samples_x[i]):
        x_samples.append(samples_x[i])
        y_samples.append(samples_y[i])
```

####(b) Plot!

```python
hist(x_samples,50,normed=True,label="Sampled Distribution")
plot(x, f(x),linewidth=2, label="True Distribution")
#plot(x, F(x),linewidth=2, label="CDF Function")   
legend(loc = 'upper left')
show()
```
