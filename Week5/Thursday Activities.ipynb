### Hit of Miss MC Integration

I. Define the function we want to integrate (it could also be an array of points)

II. Create the bounding box around our function within which to work

III. Randomly (or selectively randomly) sample points within that region

IV. Calculate the fraction of points that fall within the function boundary

V.  The integral is then the point fraction times the area of the bounding box.

# Example 1

```python
import math
```

$ f(x)= x^{1/2} + sin(x) $

```python
f = lambda x: x**(0.5) + sin(x)
```

```python
x = arange(0.,15.,0.001)
```

```python
myplot = plot(x,f(x),lw=3)
ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
xlabel(r'$x$', fontsize = 20)
xlim(0.,15.)

show()
```

Suppose we want to integrate this function between the limits (1.7,12.4). We can create the bounding box (Step II) for our random sampling for these limits and the maximum value of the function within that range.

```python
xmin = 1.7
xmax = 12.4
#find the maximum value of f within the boundary to set ymax
subx = x[logical_and((x > xmin),(x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

#plot the box
plot([xmin,xmin],[ymin,ymax], color='k', linestyle='--') #vertical left
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--') #vertical right
plot([xmin, xmax],[ymax,ymax],color='k', linestyle='--') #horizontal top

show()
```

Step III is to create random sample points within our boundary. You should investigate the random library in numpy. Let's create a set of sample points along the x and y axes. The random_sample() function returns random numbers within the bounds [0.,1.). To get the values within our limits, we need to scale and translate the values into our range:

```python
#random_sample?
N=10000
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax - ymin)*random_sample(N);
#replot the funaction
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],'.')
#plot the box
plot([xmin,xmin],[ymin,ymax], color='k', linestyle='--') #vertical left
plot([xmax,xmax],[ymin,ymax], color='k',linestyle='--') #vertical right
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--') #horizontal top

show()
```

```python
newmask = (samples_y < f(samples_x))
```

```python
sum(newmask)
```

Now we can replot the points the lie below the function for proof:

```python
#replot the funaction
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],'.')
#show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'r.')

#plot the box
plot([xmin,xmin],[ymin,ymax], color='k', linestyle='--') #vertical left
plot([xmax,xmax],[ymin,ymax], color='k',linestyle='--') #vertical right
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--') #horizontal top

show()
```

Now we can compute our integral from step V: the area of the rectangle times the ratio of the number of red and green dots.

```python
I = (xmax - xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

We can evaluate this at the two endpoints to obtain the exact solution:

```python
Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax)-Iana(xmin)
Ierr = 100*(Iexact-I)/Iexact
print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
```
