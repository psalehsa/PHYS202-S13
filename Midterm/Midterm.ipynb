```python
def is_anagram(s1,s2):
    count = 0 #count to see length
    if len(s1) == len(s2): #have to make sure they are they same length or they can't have all the same letters
        #if s1[:] == s2[:]:
            #return True
        for letter in s1:
            if letter in s2:
                if count == len(s1):
                    return True
                else:
                    count += 1 #once count = length then all the letters have been accounted for
            else:
                return False
    else:
        return False
    
print is_anagram("hi","ih")
```
