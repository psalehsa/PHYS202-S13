a) Write a reusable function Temp(Resistance) that takes any resistance and returns the interpolated temperature.

```python
def finiteDifference(x,y):
    """Takes as arguements arrays of values of x and y(x) and returns the derivative of y with respect to x"""
    dy=diff(y) # y(n+1)-y(n)
    dx=diff(x)
    dydx=zeros(y.shape,float)
    dydx[1:-1]=(y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0]=(y[1]-y[0])/(x[1]-x[0])
    dydx[-1]=(y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

b) Use the above function to take the derivative of the function y=x^5 over the range 0 < x < 10, with steps of 0.1.

```python
x=arange(0.,10.,0.1)
y=x**5
```

```python
print finiteDifference(x,y)
```

```python
dydxExact=5*x**4
```

```python
plot(x,finiteDifference(x,y),label='approx dydx')
plot(x,dydxExact,label='exact dydx')
legend(loc='upper left')
show()
```

c) What is the accuracy of the result if you use a spacing of 0.001?

```python
x=arange(0.,10.,0.001)
y=x**5
plot(x,finiteDifference(x,y),label='dydx')
legend(loc='upper left')
show()
```

```python
def fourPtFiniteDiff(x,y):
    """Computes derivative using the four-point center differencing method"""
    dydx = zeros(y.shape,float)
    dydx[2:-2]=(y[3:]-y[:-3])/(x[3:]-x[:-3])
    dydx[0] = (y[1] - y[0]) / (x[1] - x[0])
    dydx[-1] = (y[-1] - y[-2]) / (x[-1] - x[-2])
    dydx[-2] = (y[-2] - y[-3]) / (x[-2] - x[-3])
    dydx[1] = (y[2] - y[1]) / (x[2] - x[1])
    return dydx

    
```

```python
x=arange(1,10,0.1)
y=x**5
dydxExact=5*x**4
dydx=fourPtFiniteDiff(x,y)
```

```python
percentError=100*abs(dydx[1:]-dydxExact[1:])/dydxExact[1:]
semilogy(x[1:],percentError,label='Function Inaccuracy')
ylabel("Percentage Error with dydxExact")
legend(loc="upper right")
show()
```
