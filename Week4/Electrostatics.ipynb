```python
from Electrostatics import *
```

```python
dx=.003 #steps of .06
dy=.003
x=arange(-.1,.1,dx)
y=arange(-.1,.1,dy)
X,Y=meshgrid(x,y)
v=pointPotential(X,Y,-2e-9,0.,-0.5e-2)+ dipolePotential(X,Y,3e-9,1.e-2)
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,v,rstride=1,cstride=1,linewidth=0,cmap=cm.gist_earth_r)
cb=fig.colorbar(p,shrink=0.5)
xlabel('x(m)')
ylabel('y(m)')
show()
```

```python
levels=array([-20000.,-10000.,-5000.,-2000.,-1000.,-500.,-250.,0.,250.,500.,1000.,2000.,5000.,10000.,20000.])
contour(X,Y,v,levels,cmap=cm.coolwarm)
xlabel('x(m)')
ylabel('y(m)')
show()
```
