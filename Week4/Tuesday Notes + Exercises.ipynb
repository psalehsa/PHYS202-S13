3D Plotting and Functions
=========================

The dipole potential is just the cum of the potentials from two point charges separated by some distance. We'll consider a pair of point charges +- located at x = 0, y = +- d.


$ V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2+(y+d)^2}}$

Recall that k=1/4$\pi$ epsilon_0. To see what is happening we will just let q=1 and d = 1. The answer won't depend in an important way on these choices.


```python
#make the x-y grid to cover the plane -2<x<2 and -2<y<2.
dx=.06 #steps of .06
dy=.06
x=arange(-2.,2.,dx)
y=arange(-2.,2.,dy)
X,Y=meshgrid(x,y)
#make the function, V(x,y)
k=1; q=1; d=1
V=(k*q/(X**2+(Y-d)**2)**(1/2.))- (k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.gist_earth_r)
cb=fig.colorbar(p,shrink=0.5)
```

```python
#Represent 3D data by contours
levels = array([8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-4,-8])
contour(X,Y,V,levels,cmap=cm.coolwarm)
show()
```

$G(x,y)=e^{-x^2-y^2}$ over the range -6 < (x,y) < 6 using 51 bins in $x$ and 51 bines in $y$.

```python
def pointPotential(x,y,q,posx,posy):
    k=8987551787
    Vxy= (k*q)/((((x-posx)**(2.))+((y-posy)**(2.)))**(1/2.))
    """Takes as arguements the point or points in space where you want to know the potential, the size(and sign) of the charge q and the position of the charge in the plane, and return the electric potential"""
    return Vxy
```

```python
def dipolePotential(x,y,q,d):
    k=8987551787
    Vxy=pointPotential(x,y,q,-d,0.)-pointPotential(x,y,q,d,0.)

    """Takes as arguements the point or points you want to know the potential, the size of the charges q, and the distince d that seperates the charges."""
    return Vxy
```
